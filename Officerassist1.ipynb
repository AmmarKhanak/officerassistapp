```python
import google.generativeai as genai
genai.configure(api_key='AIzaSyCKmGuz343tl8Dgn_Rlnksu99Qak3PF4hU')
model = genai.GenerativeModel('gemini-flash-latest')
print("ai is ready to work")

# 1. Upload the audio file to the AI
audio_file = genai.upload_file(path='report.mp3.m4a')

print("Processing audio...")

# 2. Create the "Prompt" (The Instructions for the AI)
prompt = """
You are an AI assistant for a Police Officer.
Listen to the attached audio recording of an incident.
Please generate a formal Police Incident Report.
The report must include:
- Incident Summary
- Key People Involved
- Timeline of Events
- Actions Taken
Keep the tone professional, objective, and factual.
"""

# 3. Send the audio and the prompt to the AI
response = model.generate_content([prompt, audio_file])

# 4. Print the result
print("--- GENERATED REPORT ---\n")
print(response.text)
```
